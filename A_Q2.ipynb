```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def _plot(img, title='image', figsize=(10,10)):
    plt.figure(figsize=figsize)
    plt.imshow(img, 'gray'), plt.title(title)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
# Read source image.
img1_ = cv2.imread("HW4\\1-1.JPG")
img2_ = cv2.imread("HW4\\1-2.JPG")
```

```python
# RGB to Gray
img1 = cv2.cvtColor(img1_, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2_, cv2.COLOR_BGR2GRAY)
```

# Image I

```python
# Four corners of the mosaic in source image
pts_src = np.array([[1176, 1057], [1371, 1094], [1328, 1184],[1115, 1141]])
pts = pts_src.reshape((-1,1,2))
im = cv2.polylines(img1,np.int32([pts]),True,(0,255,255))
_plot(im, 'Source Image + PolyLine')
```

```python
# Four corners of the masaic in destination image.
pts_dst = np.array([[1200, 800], [1400, 800], [1400, 1000],[1200, 1000]])

# Calculate Homography
h, status = cv2.findHomography(pts_src, pts_dst)

# Warp source image to destination based on homography
im_out = cv2.warpPerspective(img1, h, (img1.shape[1],img1.shape[0]))
im_out1 = im_out[:1400,400:2100]

# Display images
_plot(im_out, 'WarpPerspective Image')
_plot(im_out1, 'Croped WarpPerspective Image')
```

# Image II

```python
# Read source image.
img2 = cv2.cvtColor(img2_, cv2.COLOR_BGR2GRAY)
# Four corners of the book in source image
pts_src = np.array([[873, 1087], [1077, 1051], [1135, 1143],[911, 1187]])
pts = pts_src.reshape((-1,1,2))
im = cv2.polylines(img2,np.int32([pts]),True,(0,255,255))
_plot(im, 'Source Image + PolyLine')
```

```python
# Four corners of the mosaics in destination image.
pts_dst = np.array([[1200, 800], [1400, 800], [1400, 1000],[1200, 1000]])

# Calculate Homography
h, status = cv2.findHomography(pts_src, pts_dst)

# Warp source image to destination based on homography
im_out = cv2.warpPerspective(img2, h, (img2.shape[1],img2.shape[0]))
im_out1 = im_out[0:1300,700:2200]

# Display images
_plot(im_out, 'WarpPerspective Image')
_plot(im_out1, 'Croped WarpPerspective Image')
```
